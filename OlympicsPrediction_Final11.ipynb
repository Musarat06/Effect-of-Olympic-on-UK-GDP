```python
columnNames = ['Year', 'GDP', 'ExchangeRate', 'CPI', 'realGDP', 'realGrowthRate', 'RetailTrade', 'manufacturingProd']

regressionData = pd.DataFrame(columns = columnNames)

regressionData['Year'] = df['Year'][2:].values
regressionData['GDP'] = df['United Kingdom'][:-2].values
regressionData['ExchangeRate'] = np.asarray(UKExchageMean)
regressionData['CPI'] = np.asarray(dfCPIUK)
regressionData['realGDP'] = np.asarray(realGDPavg)
regressionData['realGrowthRate'] = np.asarray(realGrowthRateavg)
regressionData['RetailTrade'] = np.asarray(RetailTradeIndex)
regressionData['manufacturingProd'] = np.asarray(ProdcutionManufacturing)
```

```python
regressionDataTrain = regressionData[0:10]
regressionDataTest = regressionData[11:]
```

```python
len(regressionDataTest)
```

# Comparison of Actual and Prediction Values of GDPs


Ordinary Least Square Regression Feature:

    * Dependent Variable (Response Variable:) Real GDP
    * Independent Variable (Explanatory Variable): Retail Trade, CPI ExchangeRate, and manufacturingProd of the UK.

This method will be used to plot a model of predicted values of GDP against the actual values of GDP. The purpose of such model is to know whether hosting Olympics increase the totall GDP or not. This will also help the countries to decided for future to host or reject to host an event like Olympics. 

```python
# models to invistigate
m1 = str('realGDP ~ RetailTrade + CPI+ ExchangeRate +manufacturingProd') #Single Regression
```

```python
train_m1_fit = smf.ols(m1,data= regressionDataTrain).fit()
```

```python
plt.figure(figsize = (15,4))
plt.subplot(1,2,1)
plt.plot( regressionData['Year'][0:10],regressionData['realGDP'][0:10], 'g--')
plt.plot( regressionData['Year'][10],regressionData['realGDP'][10], 'b*')
plt.plot( regressionData['Year'][11:],regressionData['realGDP'][11:], 'r--')
plt.xticks(regressionData['Year'][0:19:3])
plt.xlabel("Year (2002-2020)")
plt.ylabel("GDP per capita")
plt.title("GDP per Capita of UK from 2002 to 2020 ")

plt.plot( regressionData['Year'][0:10],train_m1_fit.fittedvalues, 'k--')
plt.plot( regressionData['Year'][10],regressionData['realGDP'][10], 'b*')
plt.plot( regressionData['Year'][11:],train_m1_fit.predict(regressionDataTest), 'y--')
plt.xticks(regressionData['Year'][0:19:3])
plt.xlabel("Year (2002-2020)")
plt.ylabel("Real GDP")
plt.title("Real GDP of UK Predicted (black and yelllow)\n vs Actual (green and red) from 2002 to 2020")

plt.show()
```

This is the result of Ordinary Least Square Regression model for United kingdom that we were looking for. There are prediction line of yellow color against the actual values of the United kingdom. This type of model will tell us whether it is better for the country to host an Olympics or not. Again the predicted GDP is slighlty goes higher than the actuall GDP's values. Our predicted values is based on the pre-olympic data of the United kingdom. The predicted values shows a slight increase in the GDP after the olympics. This simple means that the GDP of united kingdom would have better if they would have not hosted Olympics.
